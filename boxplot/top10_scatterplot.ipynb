```python
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.options.display.max_columns = 70
pd.options.display.max_rows = 20

top10_data = pd.read_csv('top_10.csv', header = None)
top10_data = top10_data.transpose()

top10_data.columns = top10_data.iloc[0] # set column header
top10_data = top10_data.drop(0) # drop duplicated row
top10_data = top10_data.drop(1) # drop duplicated row
top10_data = top10_data.drop(8) # drop duplicated row
top10_data['cat'] = "top 10" # create top/overall cat
top10_data.rename(columns={'industry':'thermometer'}, inplace=True)

top100_data = pd.read_csv('top_100.csv', header = None)
top100_data = top100_data.transpose()

top100_data.columns = top100_data.iloc[0] # set column header
top100_data = top100_data.drop(0) # drop duplicated row
top100_data = top100_data.drop(1) # drop duplicated row
top100_data = top100_data.drop(8) # drop duplicated row
top100_data['cat'] = "top 100" # create top/overall cat
top100_data.rename(columns={'industry':'thermometer'}, inplace=True)

top10_data
```

```python
import matplotlib.pyplot as plt

prev = None # print unique columns

for column in top10_data:
    if column != "cat" and column != "thermometer" and column != prev:
        print column
        prev = column
        temp = pd.melt(top10_data, id_vars=['thermometer', 'cat'], value_vars=[column])
        temp100 = pd.melt(top100_data, id_vars=['thermometer', 'cat'], value_vars=[column])
        result = temp.append(temp100)
        result['value'] = result['value'].astype(float)
        sns.set()
        sns.set_context("paper")
        sns.stripplot(data = result, x="thermometer", y='value', hue='cat')
        plt.show()
```
