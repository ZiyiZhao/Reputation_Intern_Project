```python
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.options.display.max_columns = 70
pd.options.display.max_rows = 20

top10_data = pd.read_csv('top_avgs.csv', header = None)
top10_data = top10_data.transpose()

top10_data.columns = top10_data.iloc[0] # set column header
top10_data = top10_data.drop(0) # drop duplicated row
top10_data['cat'] = "top" # create top/overall cat

top10_data.rename(columns={'industry':'thermometer'}, inplace=True)

top10_data
```

```python
overall_data = pd.read_csv('overall_avgs.csv', header = None)
overall_data = overall_data.transpose()

overall_data.columns = overall_data.iloc[0] # set column header
overall_data = overall_data.drop(0)
overall_data['cat'] = "overall"

overall_data.rename(columns={'industry':'thermometer'}, inplace=True)

overall_data
```

```python
result = top10_data.append(overall_data)

result
```

```python
import matplotlib.pyplot as plt

for column in result:
    if column != "cat" and column != "thermometer":
        print column
        result[column] = pd.to_numeric(result[column], errors='coerce')
        sns.set()
        sns.set_context("paper")
        sns.barplot(data = result, x="thermometer", y=column, hue="cat")
        plt.show()
```
